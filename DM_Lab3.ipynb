#### Question 1. Visualization the column price in diamonds dataset with the following requirements:


A. Create equally interval bins by min and max of price

B. Plot the histogram of each bins using hist and countplot

C. Show the percent on the top of column in each interval

```python
nbins = 17
arr = np.array(dmd.price)
bins = np.linspace(arr.min(), arr.max(), nbins + 1)
print(f"Number of interval {nbins} bins is calculated : ")
display(bins)
dmd['binned'] = pd.cut(arr, bins)
display(dmd.head())
print(dmd["binned"].value_counts())
```

```python
bins
```

```python
figure(figsize=(20, 8))
plt.title("Histogram of Price", size = 15)
plt.xlabel("Bin of Price", size = 15)
plt.ylabel("Frequency of each bins", size = 15)
density , bins, _ = plt.hist(dmd.price, bins = 17, alpha=0.5, color = "g")
for x,y in zip(bins, density):
    if y != 0:
        plt.text(x, y+0.05, y, fontsize=15) # x,y,str 
plt.show()
```

```python
figure(figsize=(20, 9))
plt.title("Count Plot of Price by Bins", size = 15)
ax = sns.countplot(x = "binned", data = dmd, alpha = 0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 40, ha = "right")
plt.xlabel("Bin of Price", size = 15)
plt.ylabel("Count of each bins", size = 15)

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(dmd))
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')

plt.show()
```

#### Question 2: Visualization the column Firing Rate in dots dataset with the following requirements:

A. Comparision of Firing Rate of T1 and T2

B. Comparison of Firing Rate and Align by Choice 

C. Distribution of Coherence and Firing Rate by Time

```python
dot1 = dot[dot['choice'] == "T1"]
display(dot1.head())
dot2 = dot[dot['choice'] == "T2"]
dot2.head()
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(12)
fig.set_figheight(10)
fig.suptitle("Bi-Variate Analysis")

sns.histplot(dot1['firing_rate'], ax=axes[0, 0], kde = True, color = 'r')
sns.histplot(dot2['firing_rate'], ax=axes[0, 0], kde = True , color = 'b'); 
axes[0, 0].set_title('T1 And T2 Firing Rate')

ax_rowspan = plt.subplot2grid((2, 2), (0, 1), rowspan=2)
sns.boxplot(x="align", y="firing_rate", hue="choice", data=dot, palette="Set2", ax=ax_rowspan)
ax_rowspan.set_title('Box plot of Align and Firing Rate')

sns.scatterplot(data=dot, x="firing_rate", y="coherence", hue="time", ax=axes[1, 0])
axes[1, 0].set_title('Distribution Coherence and Firing Rate')

plt.show()
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(20)
fig.set_figheight(15)
fig.suptitle("Bi-Variate Analysis")

ax_colspan = plt.subplot2grid((2, 2), (0, 0), colspan=2)
sns.histplot(dot1.firing_rate, kde = True, color = 'r', ax= ax_colspan)
sns.histplot(dot2.firing_rate, kde = True , color = 'b', ax= ax_colspan); 
ax_colspan.set_title('T1 and T2 Firing Rate')

sns.boxplot(x="align", y="firing_rate", hue="choice", data=dot, palette="Set3", ax=axes[1, 1])
axes[1, 1].set_title('Box plot of Align and Firing Rate')

sns.scatterplot(data=dot, x="firing_rate", y="coherence", hue="time", ax=axes[1, 0])
axes[1, 0].set_title('Distribution Coherence and Firing Rate')

plt.show() 
```
